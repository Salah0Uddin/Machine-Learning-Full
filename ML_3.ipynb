#Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
iris
```

```python
X=df.iloc[:,:-1]
y=iris.target
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
#by default post pruning will occur
dtclassifier=DecisionTreeClassifier()
```

```python
dtclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier,filled=True)
```

**As we can see there are many branches with not necessary leaf nodes so now we perform post pruning to cut them down and reduce the overfitting**

```python
dtclassifier2=DecisionTreeClassifier(max_depth=2)
```

```python
dtclassifier2.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,8))
tree.plot_tree(dtclassifier2,filled=True)
```

```python
y_pred=dtclassifier2.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
