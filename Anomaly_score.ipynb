# Isolation Forest Anomaly Detection Score


```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/content/healthcare.csv')
df.head()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
clf=IsolationForest(contamination=0.2)  # insted of 0.2 use 'auto' it is highly recommended
clf.fit(df)
predictions=clf.predict(df)
```

```python
predictions
```

```python
import numpy as np
index = np.where(predictions < 0)
index
```

```python
x=df.values
x
```

```python
index=np.where(predictions < 0)
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors="r")
```
